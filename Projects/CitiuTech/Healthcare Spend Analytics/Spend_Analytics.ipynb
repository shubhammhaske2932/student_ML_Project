```python
df = pd.read_csv("Final_Dataset_V2.csv")
df.head()
```

```python
df['Item_Description'].value_counts()
```

```python
def Item(x):
    if x == 'Face Mask':
        return 'PK100'
    elif x == 'Gloves':
        return 'PK100'
    elif x == 'Surgical Mask':
        return 'PK100'
    elif x == 'Cotton Swabs':
        return 'PK100'
    elif x == 'Bandage':
        return 'PK50'
    elif x == 'Syringe':
        return 'PK50'
    elif x == 'Oxygen Mask':
        return 'PK50'
    elif x == 'Needle':
        return 'PK50'
    elif x == 'Needle Holder':
        return 'PK50'
    elif x == 'Forceps':
        return 'DZ'
    elif x == 'Cotton Wool':
        return 'DZ'
    elif x == 'Scissor':
        return 'DZ'
    elif x == 'Saline Bag':
        return 'DZ'
    else:
        return 'EA'
```

```python
df['Lowest_Unit_Of_Measure'] = df['Item_Description'].apply(Item)
```

```python
df['Purchasing_Unit_Of_Measure'] = df['Item_Description'].apply(Item)
```

```python
df['Unit_Of_Measure'] = df['Item_Description'].apply(Item)
```

```python
def UOM_NUM(x):
    if x == 'DZ':
        return 12
    elif x == 'PK100':
        return 100
    elif x == 'PK50':
        return 50
    else:
        return 1
```

```python
df['Unit_Of_Measure_In_Number'] = df['Unit_Of_Measure'].apply(UOM_NUM)
```

```python
df.head(10)
```

```python
df = df.rename({"Unit_Price ($)\t " :"Unit_Price"})
```

```python
df.dtypes
```

```python
df['Unit_Price'] = df['Unit_Price ($)\t']
```

```python
df.drop("Unit_Price ($)\t",axis=1,inplace = True)
```

```python
df.head()
```

```python
df['Price_Per_UOM'] = df['Unit_Of_Measure_In_Number'] * df['Unit_Price']
```

```python
df.head()
```

```python
df['Total_Price ($)'] = df['Quantity_(at_UOM)'] * df['Price_Per_UOM']
```

```python
df.drop('Unit_Price', axis=1, inplace = True)
```
